# Manipulando o Dataframe
#### Selecionando dados

```python
# Podemos acessar informações dos dataframes e series de diversas maneiras
# Quando queremos apenas 1 coluna

df['Idade']
```

```python
# Quando queremos mais de uma coluna

df[['Idade','dia','mes']]
```

# iLoc - Seleção

```python
df.iloc[[1]]
```

Seleção de Linhas

```python
# Apenas uma linha

df.iloc[[1]]
```

```python
# Apenas uma linha

df.iloc[[-3]]
```

```python
# Apenas uma linha
df.iloc[1:3]
```

```python
# Seleção de Colunas - Apenas uma coluna

df.iloc[:,1]
```

```python
# Seleção de Colunas - Fatiamento de colunas

df.iloc[:,1:3]
```

```python
# Seleção de linhas e colunas - Um fatiamento de linha e colunas

df.iloc[0:2,0:3]
```

```python
# Um array e um fatiamento de colunas

df.iloc[[1,2,400],0:3]
```

Seleção Loc

```python
# Selecionando apenas valores de uma linha

df.loc[[1]]
```

```python
df.loc[1]
```

```python
# Selecionando uma lista de linhas

df.loc[[0,1,30]]
```

```python
# Selecionando uma fatia de dados

df.loc[0:2]
```

```python
# Selecionando uma lista de colunas

df.loc[[0,1],['Idade','Profissao','default']]
```

```python
# Selecionando uma fatia de linhas e uma lista de colunas

df.loc[0:10,['Idade','Profissao','default']]
```

```python
# Selecionando com base em uma condição ou mais condições

df.loc[(df.Profissao == "management") & (df.Idade < 30) | (df.educacao =="secondary") ]
```
