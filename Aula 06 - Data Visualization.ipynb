```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 6)
y = np.arange(2, 11, 2)

plt.plot(x, y)
```

```python
plt.plot(x, y, marker="*")
```

```python
plt.plot(x, y, marker=".", c="red")
```

```python
plt.plot(x, y, marker="*", c="red", ds="steps", ls="--")
```

```python
plt.plot(x, y, marker="*", c="red", ds="steps", ls="--", mec="#0012d2")
```

```python
plt.scatter(x, y)
plt.xlabel("Valores de X", fontsize=20)
plt.ylabel("Valores de Y", fontsize=15)

plt.title("Nome do Grfico", c="purple")
plt.xticks(fontsize=15, c="red")
plt.yticks(fontsize=17, c="green")
```

```python
sns.pairplot(data=iris, hue="species", diag_kind="kde", kind="scatter", palette="husl")
plt.savefig("pairplot_iris.png")
```

```python
sns.displot(x=iris["petal_length"], kde=True)
```

```python
sns.displot(x=iris["petal_length"], kde=False)
```

```python
sns.displot(x=iris["petal_length"], kind="kde")
```

```python
sns.set(rc={"figure.figsize": (16, 6)})

plt.subplot(1, 2, 1)
plt.title("stripplot", fontsize=18)
sns.stripplot(data=iris, x="species", y="petal_length")

plt.subplot(1, 2, 2)
plt.title("swarmplot", fontsize=18)
sns.swarmplot(data=iris, x="species", y="petal_length")
```

```python
sns.set(rc={"figure.figsize": (16, 6)})


plt.title("Alphas")

plt.subplot(1, 3, 1)
plt.title("alpha = 1")
sns.stripplot(data=iris, x="species", y="petal_length")

plt.subplot(1, 3, 2)
plt.title("alpha = 0.75")
sns.stripplot(data=iris, x="species", y="petal_length", alpha=0.6)

plt.subplot(1, 3, 3)
plt.title("alpha = 0.5")
sns.stripplot(data=iris, x="species", y="petal_length", alpha=0.3)
```

```python
sns.barplot(data=iris, x="species", y="petal_length")
```
