1. Abra o arquivo preferencias.csv como um dataframe chamado 
pref.

```python
import pandas as pd
```

```python
pref = pd.read_csv('preferencias.csv')
pref
```

```python
pref.rename (
    columns={
        'Favorite Color':'favoriteColor',
        'Favorite Music Genre':'favoriteMusicGenre',
        'Favorite Beverage':'favoriteBeverage',
        'Favorite Soft Drink':'favoriteSoftDrink',
        'Gender':'gender'
    },inplace=True
)

pref
```

3. Agrupe os dados pela variável Gender (gênero).

```python
pref.groupby('gender') ## Agrupando pela variável Gender
```

```python
pref.groupby('gender').groups ## Visualizando os Grupos
```

```python
pref.groupby('gender').count()
```
